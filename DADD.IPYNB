```python
import pandas as pd

# CSV 파일 로드
file_path = 'C:/Users/NT551/Desktop/2024.06.13  케라스/diabetes_data_upload.csv'  # 실제 파일 경로로 변경
df = pd.read_csv(file_path)

# 데이터 변환 함수 정의
def yes_no_to_binary(value):
    if str(value).strip().upper() == "YES":
        return 1
    elif str(value).strip().upper() == "NO":
        return 0
    return value

# 성별 변환 함수 정의
def gender_to_numeric(value):
    if str(value).strip().capitalize() == "Male":
        return 1
    elif str(value).strip().capitalize() == "Female":
        return 2
    return value

# class 변환 함수 정의
def class_to_binary(value):
    if str(value).strip().capitalize() == "Positive":
        return 1
    elif str(value).strip().capitalize() == "Negative":
        return 0
    return value

# YES/NO 변환할 열 목록
columns_to_convert = [
    'Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 
    'Polyphagia', 'Genital thrush', 'visual blurring', 'Itching', 
    'Irritability', 'delayed healing', 'partial paresis', 
    'muscle stiffness', 'Alopecia', 'Obesity'
]

# 각 열에 변환 함수 적용
for column in columns_to_convert:
    df[column] = df[column].apply(yes_no_to_binary)

# 성별 변환
df['Gender'] = df['Gender'].apply(gender_to_numeric)

# class 변환
df['class'] = df['class'].apply(class_to_binary)

# 결과 확인
print(df.head())

# 변환된 데이터 저장
output_file_path = 'converted_diabetes_data.xlsx'
df.to_excel(output_file_path, index=False)

print(f"변환된 파일이 {output_file_path}에 저장되었습니다.")
```

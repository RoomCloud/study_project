```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'C:/Users/NT551/Desktop/데이터 추출(연습)/last_row_of_collected_datas.xlsx'
data = pd.read_excel(file_path)
```

```python
# 챔피언 선택 데이터 추출
champion_columns = [f'championName_{i}' for i in range(10)]
champion_choices = data[champion_columns].values.flatten()



# 챔피언 빈도수 계산
champion_counts = pd.Series(champion_choices).value_counts()

# 상위 10% 챔피언 선택
top_15_champions = champion_counts.head(15)

# 데이터 프레임으로 변환 (선택 사항)
champion_counts_df = champion_counts.reset_index()
champion_counts_df.columns = ['Champion', 'Count']

# 시각화
plt.figure(figsize=(18, 10))  # 그래프 크기 조정
sns.set(style="whitegrid")
plt.barh(champion_counts_df['Champion'], champion_counts_df['Count'], color='skyblue')
plt.xlabel('Count', fontsize=14)  # x축 레이블 크기 조정
plt.ylabel('Champion', fontsize=14)  # y축 레이블 크기 조정
plt.title('Frequency of Most Common Champions', fontsize=16)  # 제목 크기 조정
plt.xticks(fontsize=12)  # x축 눈금 글씨 크기 조정
plt.yticks(fontsize=12)  # y축 눈금 글씨 크기 조정
plt.gca().invert_yaxis()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 엑셀 파일 불러오기
file_path = 'C:/Users/NT551/Desktop/데이터 추출(연습)/last_row_of_collected_datas.xlsx'
data = pd.read_excel(file_path)

# 챔피언 선택 데이터 추출
champion_columns = [f'championName_{i}' for i in range(10)]
champion_choices = data[champion_columns].values.flatten()

# 챔피언 빈도수 계산
champion_counts = pd.Series(champion_choices).value_counts()

# 상위 15개 챔피언 선택
#top_15_champions = champion_counts.head(15)
top_10_percent_threshold = champion_counts.sum() * 0.10
top_champions = champion_counts[champion_counts.cumsum() <= top_10_percent_threshold]

# 데이터 프레임으로 변환 (선택 사항)
# top_15_champions_df = top_15_champions.reset_index()
# top_15_champions_df.columns = ['Champion', 'Count']
top_champions_df = top_champions.reset_index()
top_champions_df.columns = ['Champion', 'Count']

# 시각화
plt.figure(figsize=(18, 10))  # 그래프 크기 조정
sns.set(style="whitegrid")
plt.barh(top_champions_df['Champion'], top_champions_df['Count'], color='skyblue')
plt.xlabel('Count', fontsize=14)  # x축 레이블 크기 조정
plt.ylabel('Champion', fontsize=14)  # y축 레이블 크기 조정
plt.title('Top 15 Most Common Champions', fontsize=16)  # 제목 크기 조정
plt.xticks(fontsize=12)  # x축 눈금 글씨 크기 조정
plt.yticks(fontsize=12)  # y축 눈금 글씨 크기 조정
plt.gca().invert_yaxis()
plt.show()
```
